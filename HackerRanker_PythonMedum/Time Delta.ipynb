## Time Delta

When users post an update on social media,such as a URL, image, status update etc., other users in their network are able to view this new post on their news feed. Users can also see exactly when the post was published, i.e, how many hours, minutes or seconds ago.

Since sometimes posts are published and viewed in different time zones, this can be confusing. You are given two timestamps of one such post that a user can see on his newsfeed in the following format:

Day dd Mon yyyy hh:mm:ss +xxxx

Here +xxxx represents the time zone. Your task is to print the absolute difference (in seconds) between them.

### Input Format

The first line contains **T**, the number of testcases.
Each testcase contains **2** lines, representing time **t_1** and time **t_2**.

### Constraints

* Input contains only valid timestamps
* **year <= 3000**.

### Output Format

Print the absolute difference **(t_1-t_2)** in seconds.

### Sample Input 0

2

Sun 10 May 2015 13:54:36 -0700

Sun 10 May 2015 13:54:36 -0000

Sat 02 May 2015 19:54:36 +0530

Fri 01 May 2015 13:54:36 -0000

### Sample Output 0

25200

88200

### Explanation 0

In the first query, when we compare the time in UTC for both the time stamps, we see a difference of 7 hours. which is **7x3600** seconds or **25200** seconds.

Similarly, in the second query, time difference is 5 hours and 30 minutes for time zone adjusting for that we have a difference of 1 day and 30 minutes. Or **24x3600+30x60 => 88200**

```python
def time_delta1(N, t1, t2):
    
    result = []
    monList = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
        
    for n in range(0, N):
        
        t1List = list(t1[n])
        d1 = int(''.join(t1List[4:6]))
        o1 = str(''.join(t1List[7:10]))
        y1 = int(''.join(t1List[11:15]))
        h1 = int(''.join(t1List[16:18]))
        m1 = int(''.join(t1List[19:21]))
        s1 = int(''.join(t1List[22:24]))
        z1 = str(''.join(t1List[25:26]))
        n1 = int(''.join(t1List[26:30]))
        
        #print('d:', d1, 'mon:', o1, 'y:', y1, 'h:', h1, 'm:', m1, 's:', s1, 'z:', z1, 'n:', n1)
        
        t2List = list(t2[n])
        d2 = int(''.join(t2List[4:6]))
        o2 = str(''.join(t2List[7:10]))
        y2 = int(''.join(t2List[11:15]))
        h2 = int(''.join(t2List[16:18]))
        m2 = int(''.join(t2List[19:21]))
        s2 = int(''.join(t2List[22:24]))
        z2 = str(''.join(t2List[25:26]))
        n2 = int(''.join(t2List[26:30]))
        
        #print('d:', d2, 'mon:', o2, 'y:', y2, 'h:', h2, 'm:', m2, 's:', s2, 'z:', z2, 'n:', n2)
        
        for mon in range(0, len(monList)):
            if o1 == monList[mon]:
                o1 = str(mon+1)
            if o2 == monList[mon]:
                o2 = str(mon+1)    
            
        yrdif = y2 - y1
        modif = int(o2) - int(o1)
        
        time_t1 = 1000000 * d1 + 10000 * h1 + 100 * m1 + 1 * s1
        time_t2 = 1000000 * d2 + 10000 * h2 + 100 * m2 + 1 * s2
       
        #print(time_t1, ' ', time_t2)
        
        time_diff = time_t2 - time_t1
        
        #print(time_diff)
        
        da_diff = abs(time_diff) // 1000000
        hr_diff = (abs(time_diff) % 1000000) // 10000
        min_diff = ((abs(time_diff) % 1000000) % 10000) // 100
        sec_diff = ((abs(time_diff) % 1000000) % 10000) % 100
    
        #print(da_diff, ' ', hr_diff, ' ', min_diff, ' ', sec_diff)
    
        if z1 == '-':
            n1 *= (-1)
        if z2 == '-':
            n2 *= (-1)
    
        n_diff = n2 - n1
        #print(n_diff)
        
        hrn_diff = abs(n_diff) // 100
        minn_diff = abs(n_diff) % 100
        
        #print(hrn_diff, ' ', minn_diff)
        
        hrs_diff = 0
        mins_diff = 0 
        
        if time_diff * n_diff > 0:
            hrs_diff = hr_diff - hrn_diff
            mins_diff = min_diff - minn_diff
        else:
            hrs_diff = hr_diff + hrn_diff
            mins_diff = min_diff + minn_diff
        
        #print(hrs_diff, ' ', mins_diff)
        
        time_diff = da_diff * 24 * 60 * 60 + hrs_diff * 60 * 60 + mins_diff * 60
        
        result.append(time_diff)
    
    return result
```

```python
def time_delta2(N, t1, t2):
    
    result = []
    monList = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    
    for n in range(0, N):
        
        t1List = list(t1[n])
        d1 = int(''.join(t1List[4:6]))
        o1 = str(''.join(t1List[7:10]))
        y1 = int(''.join(t1List[11:15]))
        h1 = int(''.join(t1List[16:18]))
        m1 = int(''.join(t1List[19:21]))
        s1 = int(''.join(t1List[22:24]))
        z1 = str(''.join(t1List[25:26]))
        n1 = int(''.join(t1List[26:30]))
        
        #print('d:', d1, 'mon:', o1, 'y:', y1, 'h:', h1, 'm:', m1, 's:', s1, 'z:', z1, 'n:', n1)
        
        t2List = list(t2[n])
        d2 = int(''.join(t2List[4:6]))
        o2 = str(''.join(t2List[7:10]))
        y2 = int(''.join(t2List[11:15]))
        h2 = int(''.join(t2List[16:18]))
        m2 = int(''.join(t2List[19:21]))
        s2 = int(''.join(t2List[22:24]))
        z2 = str(''.join(t2List[25:26]))
        n2 = int(''.join(t2List[26:30]))
        
        #print('d:', d2, 'mon:', o2, 'y:', y2, 'h:', h2, 'm:', m2, 's:', s2, 'z:', z2, 'n:', n2)
        
        for mon in range(0, len(monList)):
            if o1 == monList[mon]:
                o1 = str(mon+1)
            if o2 == monList[mon]:
                o2 = str(mon+1)    
        
        time_diff = abs(y2 - y1) * 12 * 27 * 24 * 60 *60
        if int(o2) - int(o1) < 0:
            time_diff += (int(o2) - int(o1) + 12) * 27 * 24 * 60 * 60
        else:
            time_diff += (int(o2) - int(o1)) * 27 * 24 * 60 * 60
        
        result.append(time_diff)
        
    return result
```
